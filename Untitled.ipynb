```python
retail["reg_date"]= pd.to_datetime(retail["reg_date"])
retail["date_sold"] = pd.to_datetime(retail["date_sold"])
retail = retail.drop("title",axis=1)
```

```python
retail = retail.dropna()
retail = pd.get_dummies(retail)
```

```python
retail
```

```python
from pmdarima.arima import auto_arima
to_train = retail[["date_sold","price"]]
```

```python
to_train
```

```python
to_trai = to_train.set_index("date_sold")
```

```python
to_trai
```

```python
to_trai["ave price"] = to_trai["price"].avg()
```

```python
to_trai = to_trai.groupby(level=0)
```

```python
to_trai
```

```python
group = to_trai.agg({'price':'mean'})
```

```python
group
```

```python
train = group[:170]
test = group[170:]
model = auto_arima(train, trace=True, error_action='ignore', suppress_warnings=True)
```

```python
model.fit(train)
```

```python
model.predict(n_periods=170)
```
